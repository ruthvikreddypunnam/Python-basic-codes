1.Read mobile dataset using pandas, and preprocess it (fill null values).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('mobile_cleaned.csv')
```

```python
d
```

```python
d.isnull().sum()
```

```python
d['weight'].fillna(d['weight'].mean(),inplace=True)
d['stand_by_time'].fillna(d['stand_by_time'].mean(),inplace=True)
d['no_of_reviews_in_gsmarena_in_week'].fillna(d['no_of_reviews_in_gsmarena_in_week'].mean(),inplace=True)
d['phone_height'].fillna(d['phone_height'].mean(),inplace=True)
d['talk_time'].fillna(d['talk_time'].mean(),inplace=True)
d['video_resolution'].fillna(d['video_resolution'].mean(),inplace=True)
d['primary_camera_resolution'].fillna(d['primary_camera_resolution'].mean(),inplace=True)
```

```python
d.isnull().sum()
```

2.Draw correlation graph for mobile price with all attributes.

correlation graphs

```python
for col in d.columns:
    if col!='price':
        x=list(d[col])
        y=list(d['price'])
        plt.scatter(x,y)
        plt.title("price correlation with "+col)
        plt.show()
```

### Observations
Price depends on sim_type,processor frequency,aperture,screen pixel density,version,scrren size,battery capacity
